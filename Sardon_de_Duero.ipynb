```python
def velocidad_entrada_1 (x):
    if x.mobility__speed__value <=50 and x.location__longitude__value <-4.436:
        return ("Correcta")
    if x.mobility__speed__value >50 and x.mobility__speed__value < 71 and x.location__longitude__value > -4.45 and  x.location__longitude__value < -4.436:
        return ("Deceleracion")
    if x.mobility__speed__value >71 and x.mobility__speed__value < 91 and x.location__longitude__value < -4.4403676:
        return ("Correcta")
    return ("Peligroso")

def velocidad_entrada_2 (x):
    if x.mobility__speed__value <=50 and x.location__longitude__value >-4.429:
        return ("Correcta")
    if x.mobility__speed__value >50 and x.mobility__speed__value < 71 and x.location__longitude__value > -4.429 and  x.location__longitude__value < -4.419:
        return ("Deceleracion")
    if x.mobility__speed__value >71 and x.mobility__speed__value < 91 and x.location__longitude__value < -4.4261560:
        return ("Correcta")
    return ("Peligroso")


def velocidad_urbana (x):
    if x.mobility__speed__value <=50:
        return ("Correcta")
    if x.mobility__speed__value >50 and x.mobility__speed__value < 65:
        return ("Inadecuada")
    return ("Peligrosa")

def mes (x,month):
    if x.mobility__speed__value >-1 and month == "enero":
        return ("enero")
    if x.mobility__speed__value >-1 and month == "febrero":
        return ("febrero")
    if x.mobility__speed__value >-1 and month == "marzo":
        return ("marzo")
    if x.mobility__speed__value >-1 and month == "abril":
        return ("abril")
    if x.mobility__speed__value >-1 and month == "mayo":
        return ("mayo")
    if x.mobility__speed__value >-1 and month == "junio":
        return ("junio")
    if x.mobility__speed__value >-1 and month == "julio":
        return ("julio")
    if x.mobility__speed__value >-1 and month == "agosto":
        return ("agosto")
    if x.mobility__speed__value >-1 and month == "septiembre":
        return ("septiembre")
    if x.mobility__speed__value >-1 and month == "octubre":
        return ("octubre")
    if x.mobility__speed__value >-1 and month == "noviembre":
        return ("noviembre")
    if x.mobility__speed__value >-1 and month == "diciembre":
        return ("diciembre")
    return (Nan)

def aÃ±o (x,aÃ±o):
    if x.mobility__speed__value >-1 and aÃ±o == "2020":
        return (2020)
    if x.mobility__speed__value >-1 and aÃ±o == "2021":
        return (2021)
```
